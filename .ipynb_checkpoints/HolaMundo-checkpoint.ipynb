# 3 Intérprete de Python y ejecución de scripts

## 3.0.1 Antes de empezar
Se pueden utilizar diferentes formas para ejecutar código:
* **Línea de comandos o terminal:** Shell de Python o interactivo (`python` / `ipython`).
* **IDE (Entornos de Desarrollo Integrados):** Eclipse, PyCharm, Sublime Text, Nano, VSCode, Atom, Spyder, etc.
* **Notebooks:** Google Colab, Azure Notebooks, Jupyter.



## 3.0.2 ¿Qué es un intérprete?
* Un programa que ejecuta otros programas.
* Una capa de software entre tu código y el hardware que lo ejecuta.
* Debe estar instalado en tu ordenador para poder ejecutar código Python.
* Para la especificación de Python, existen varias implementaciones:
    * **CPython:** Implementación en C (es la más común).
    * **Jython:** Implementación en Java.
    * **IronPython:** Implementación en .NET.



## 3.0.3 ¿Cómo se lleva a cabo la ejecución de scripts?

````python
#Imprimir en pantalla hello"
print('Hello World')
````

> **Guía con detalles para Windows, Linux y macOS:** [Real Python - Run Python Scripts](https://realpython.com/run-python-scripts/)



## 3.0.4 Perspectiva del desarrollador
* **Un script en Python es un fichero de texto que:**
    * Contiene instrucciones Python.
    * Tiene extensión `.py`.
* **Puedes ejecutar scripts desde:**
    * Línea de comandos.
    * IDE (Entorno de Desarrollo Integrado).



## 3.0.5 Perspectiva de Python (Ejecución Interna)
1. **Compilación de código fuente a *byte code*:**
    * El código *byte code* se ejecuta más rápido.
    * Genera ficheros `.pyc` que se almacenan en caché.
    * Permite saltarse el paso de compilación en futuras ejecuciones.
2. **Python Virtual Machine (PVM):**
    * Ejecuta las instrucciones traducidas en *byte code*.



## 3.0.6 ¿Cómo puedes ejecutar tus scripts?

### 3.0.7 Línea de comandos
* Ejecutar `py` o `python` en la terminal para abrir una sesión interactiva del intérprete.
* También se puede abrir ejecutando la aplicación "Python" desde el menú de inicio.
* Los caracteres `>>>` indican que estás dentro de una sesión interactiva.
* Util para experimentación rápida y testing.

```python
>>> print('Hello world')
Hello world

Incovenientemente: los programas que ejecutas en la linea de comandos desaparecen tras ser ejecutados



### 3.0.8 REPL
Sistema interactivo para comunicarse con el ordenador en un leguaje, Python. Se debe cumplir: *
Read. El ordenador pueda leer unidades como entrada * Evaluate. El código pueda ser procesado
* Print. Los resultados puedan verse * Loop. Continuar con la conversación

### 3.0.9 FICHEROS
* Permite almacenar programas.
* Ficheros de texto con instrucciones Python.
    - No olvidar shebang en Linux -> #!/usr/bin/env python3
* Terminologia (varia segun fuentes):
    - Scrips o programas: programa principal.
    - Modulos: ficheros inmportados desde otros fichero.
* Se puede lanzar pasando nombre de fichero a comando python.
> python ./script1.py
Hello world

* Otra alternativa (disponible a partir de Python 3.3) es:
> py ./script1.py
Hello world

* O incluso:
> ./script1.py
Hello world
